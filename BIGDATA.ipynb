```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Function to monitor memory usage
def memory_usage_mb(df):
    return df.memory_usage(deep=True).sum() / 1024**2  # Memory usage in MB

# Merge patient data with conditions and medications
# Note: This part was already fixed in a previous response to use 'PATIENT'
patient_conditions = pd.merge(patients_df, conditions_df, left_on='Id', right_on='PATIENT', how='left')
patient_data = pd.merge(patient_conditions, medications_df, on='PATIENT', how='left')

# Print initial memory usage
print(f"Initial memory usage of patient_data: {memory_usage_mb(patient_data):.2f} MB")

# Feature engineering
# Calculate Age from BIRTHDATE (since 'Age' doesn't exist directly)
patient_data['BIRTHDATE'] = pd.to_datetime(patient_data['BIRTHDATE'])
patient_data['Age'] = (pd.Timestamp('2025-04-01') - patient_data['BIRTHDATE']).dt.days // 365  # Current date is 2025-04-01

# Label encode 'DESCRIPTION_y' (medications) instead of one-hot encoding to save memory
if 'DESCRIPTION_y' in patient_data.columns:
    le = LabelEncoder()
    patient_data['Medication_Encoded'] = le.fit_transform(patient_data['DESCRIPTION_y'].astype(str))
    patient_data = patient_data.drop(columns=['DESCRIPTION_y'])  # Drop the original column

# One-hot encode 'GENDER' (should be safe since it has few unique values)
patient_data = pd.get_dummies(patient_data, columns=['GENDER'], drop_first=True)

# Print memory usage after encoding
print(f"Memory usage after encoding: {memory_usage_mb(patient_data):.2f} MB")

# Create labels (e.g., binary labels for each disease)
# Using 'DESCRIPTION_x' column from conditions_df (not 'Condition')
disease_column = 'DESCRIPTION_x'
if disease_column not in patient_data.columns:
    print("Disease column not found. Available columns:", patient_data.columns)
else:
    # Limit to the top 50 most frequent diseases to save memory
    top_n_diseases = patient_data[disease_column].value_counts().head(50).index
    diseases = [disease for disease in top_n_diseases if pd.notna(disease)]

    # Create binary labels only for the top diseases
    for disease in diseases:
        patient_data[f'Label_{disease}'] = (patient_data[disease_column] == disease).astype('int8')  # Use int8 to save memory

# Print memory usage after label creation
print(f"Memory usage after label creation: {memory_usage_mb(patient_data):.2f} MB")

# Drop unnecessary columns and handle missing values
columns_to_drop = [disease_column, 'PATIENT', 'START', 'STOP', 'ENCOUNTER', 'CODE_x', 'CODE_y',
                   'BIRTHDATE', 'DEATHDATE', 'SSN', 'DRIVERS', 'PASSPORT', 'PREFIX', 'FIRST',
                   'LAST', 'SUFFIX', 'MAIDEN', 'MARITAL', 'RACE', 'ETHNICITY', 'BIRTHPLACE',
                   'ADDRESS', 'CITY', 'STATE', 'COUNTY', 'ZIP', 'LAT', 'LON', 'PAYER',
                   'BASE_COST', 'PAYER_COVERAGE', 'DISPENSES', 'TOTALCOST', 'REASONCODE',
                   'REASONDESCRIPTION']
columns_to_drop = [col for col in columns_to_drop if col in patient_data.columns]
patient_data = patient_data.drop(columns=columns_to_drop)

# Handle missing values
patient_data = patient_data.fillna(0)

# Optimize data types for remaining columns
for col in patient_data.columns:
    if patient_data[col].dtype == 'float64':
        patient_data[col] = patient_data[col].astype('float32')
    elif patient_data[col].dtype == 'int64':
        patient_data[col] = patient_data[col].astype('int32')

# Print final memory usage
print(f"Final memory usage of patient_data: {memory_usage_mb(patient_data):.2f} MB")

# Display the processed DataFrame
print("Processed DataFrame:")
print(patient_data.head())
```

```python
import torch
from torch_geometric.data import Data
import numpy as np
from scipy.sparse import coo_matrix
import pandas as pd
import gc
import tempfile
import os

# Force subsampling to start with a very small subset of the data
subsample_frac = 0.0001  # Use 0.01% of the data to start (e.g., 6 nodes if original is 64,530)
patient_data_sub = patient_data.sample(frac=subsample_frac, random_state=42)
print(f"Number of nodes after subsampling: {len(patient_data_sub)}")

# Debug: Check columns in patient_data_sub
print("Columns in patient_data_sub after subsampling:", patient_data_sub.columns.tolist())

# Node features (excluding label columns)
feature_cols = [col for col in patient_data_sub.columns if not col.startswith('Label_')]

# Debug: Print the data types of the feature columns
print("Feature columns and their data types:")
for col in feature_cols:
    print(f"{col}: {patient_data_sub[col].dtype}")

# Ensure all feature columns are numeric
for col in feature_cols:
    if patient_data_sub[col].dtype == 'object':
        patient_data_sub[col] = pd.to_numeric(patient_data_sub[col], errors='coerce')
        patient_data_sub[col] = patient_data_sub[col].fillna(0)
    patient_data_sub[col] = patient_data_sub[col].astype('float32')

# Convert features to tensor
x = torch.tensor(patient_data_sub[feature_cols].values, dtype=torch.float32)

# Labels (multi-task: one label per disease)
label_cols = [col for col in patient_data_sub.columns if col.startswith('Label_')]
print(f"Label columns in patient_data_sub: {label_cols}")
# Ensure label columns are numeric
for col in label_cols:
    patient_data_sub[col] = patient_data_sub[col].astype('float32')
y = torch.tensor(patient_data_sub[label_cols].values, dtype=torch.float32)

# Add minimal edge creation
# Use only the top 1 disease to minimize edges
top_n_diseases = 1

# Check if label_cols is empty
if not label_cols:
    print("No label columns found in patient_data_sub. Cannot create edges.")
    edge_index = torch.empty((2, 0), dtype=torch.long)  # Empty edge_index
else:
    # Debug: Print the sum of each label column to check for non-zero values
    label_sums = patient_data_sub[label_cols].sum()
    print("Sum of each label column in patient_data_sub:")
    print(label_sums)

    disease_frequencies = label_sums.nlargest(top_n_diseases).index
    if len(disease_frequencies) == 0:
        print("No diseases found in the subsampled data (all label columns sum to 0). Using empty edge_index.")
        edge_index = torch.empty((2, 0), dtype=torch.long)  # Empty edge_index
    else:
        disease_indices = [label_cols.index(col) for col in disease_frequencies]
        label_matrix = y.numpy()[:, disease_indices]  # Shape: [num_nodes, 1]
        print(f"Using top {top_n_diseases} disease for edge creation: {disease_frequencies.tolist()}")

        # Compute the adjacency matrix
        adj_matrix = coo_matrix(label_matrix @ label_matrix.T)

        # Extract edges and write to disk to reduce memory usage
        rows, cols = adj_matrix.row, adj_matrix.col
        edge_weights = adj_matrix.data

        # Write edges to a temporary file
        with tempfile.NamedTemporaryFile(delete=False, mode='w') as temp_file:
            edge_count = 0
            for i, j, weight in zip(rows, cols, edge_weights):
                if i < j and weight >= 1:  # Since top_n_diseases=1, min_shared_conditions=1
                    temp_file.write(f"{i},{j}\n")
                    temp_file.write(f"{j},{i}\n")
                    edge_count += 2
            temp_file_path = temp_file.name

        # Read edges from the temporary file in chunks to create the edge_index tensor
        edge_indices = []
        edge_tensor = None  # Initialize the tensor to None

        with open(temp_file_path, 'r') as f:
            for line in f:
                i, j = map(int, line.strip().split(','))
                edge_indices.append([i, j])
                if len(edge_indices) >= 1000:  # Process in chunks of 1,000 edges
                    chunk_tensor = torch.tensor(edge_indices, dtype=torch.long)
                    if edge_tensor is None:
                        edge_tensor = chunk_tensor
                    else:
                        edge_tensor = torch.cat([edge_tensor, chunk_tensor], dim=0)
                    edge_indices = []  # Clear the list to free memory

        # Handle any remaining edges
        if edge_indices:
            chunk_tensor = torch.tensor(edge_indices, dtype=torch.long)
            if edge_tensor is None:
                edge_tensor = chunk_tensor
            else:
                edge_tensor = torch.cat([edge_tensor, chunk_tensor], dim=0)

        # If no edges were found, use an empty tensor
        if edge_tensor is None:
            edge_index = torch.empty((2, 0), dtype=torch.long)
        else:
            edge_index = edge_tensor.t().contiguous()
        print(f"Number of edges: {edge_index.shape[1]}")

        # Clean up the temporary file
        os.remove(temp_file_path)

# Create or update the Data object with edges
data = Data(x=x, edge_index=edge_index, y=y)
print("Graph Data (with minimal edges):")
print(data)

# Free up memory
try:
    del label_matrix, adj_matrix, edge_indices, rows, cols, edge_weights, edge_tensor
except NameError:
    pass  # Variables were not defined, so nothing to delete
gc.collect()
```

```python
import torch
from torch_geometric.data import Data
import numpy as np
import pandas as pd
from scipy.sparse import coo_matrix
from sklearn.cluster import AgglomerativeClustering
import gc

# Assuming 'data' is the Data object from Step 3
# data.x: Node features [num_nodes, num_features]
# data.edge_index: Edge indices [2, num_edges]
# data.y: Labels [num_nodes, num_diseases]

# Step 4: Implement Dynamic Task Grouping

# 1. Compute Disease Co-occurrence Matrix
# y is the label matrix of shape [num_nodes, num_diseases]
# Compute co-occurrence as the dot product of y^T * y
y_numpy = data.y.numpy()  # Convert to numpy for computation
co_occurrence_matrix = y_numpy.T @ y_numpy  # Shape: [num_diseases, num_diseases]
# co_occurrence_matrix[i, j] represents the number of patients with both disease i and disease j

# Normalize the co-occurrence matrix to get a similarity score (e.g., cosine similarity)
# Add a small epsilon to avoid division by zero
epsilon = 1e-8
disease_magnitudes = np.sqrt(np.diag(co_occurrence_matrix))  # Square root of diagonal elements
disease_magnitudes = disease_magnitudes.reshape(-1, 1)  # Shape: [num_diseases, 1]
similarity_matrix = co_occurrence_matrix / (disease_magnitudes @ disease_magnitudes.T + epsilon)
# similarity_matrix[i, j] is the cosine similarity between disease i and disease j

# Convert similarity to distance for clustering (distance = 1 - similarity)
distance_matrix = 1 - similarity_matrix
# Ensure the distance matrix is non-negative and the diagonal is 0
distance_matrix = np.maximum(distance_matrix, 0)
np.fill_diagonal(distance_matrix, 0)

# 2. Cluster Diseases into Groups
# Use hierarchical clustering (AgglomerativeClustering) to group diseases
num_groups = min(5, len(label_cols))  # Number of groups (e.g., 5, or fewer if num_diseases is small)
if len(label_cols) > 1:  # Only cluster if there are at least 2 diseases
    clustering = AgglomerativeClustering(
        n_clusters=num_groups,
        metric='precomputed',  # Updated parameter name
        linkage='average'
    )
    cluster_labels = clustering.fit_predict(distance_matrix)
else:
    # If there's only 1 disease (or none), assign it to a single group
    cluster_labels = np.zeros(len(label_cols), dtype=int)

# 3. Create Task Groups
# Map each disease to its cluster (group)
task_groups = [[] for _ in range(num_groups)]
for disease_idx, cluster_id in enumerate(cluster_labels):
    task_groups[cluster_id].append(disease_idx)

# Print the task groups for debugging
print("Task Groups (indices of diseases in each group):")
for group_idx, group in enumerate(task_groups):
    print(f"Group {group_idx}: {group}")
    # Map indices back to disease names for better understanding
    disease_names = [label_cols[idx] for idx in group]
    print(f"Group {group_idx} diseases: {disease_names}")

# 4. Prepare Task Group Information for GNN Training
# Create a tensor mapping each disease to its group
task_to_group = torch.tensor(cluster_labels, dtype=torch.long)
# task_to_group[i] is the group ID of disease i

# Add task group information to the Data object
data.task_to_group = task_to_group
data.num_groups = num_groups
data.task_groups = task_groups  # List of lists for easy access during training

# Print the updated Data object
print("Updated Graph Data with Task Groups:")
print(data)

# Free up memory
del y_numpy, co_occurrence_matrix, similarity_matrix, distance_matrix, cluster_labels
gc.collect()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from sklearn.metrics import roc_auc_score, f1_score
import numpy as np
import gc

# Step 5: Define and Train the GNN Model with Task Grouping (Reduced Epochs)

# 1. Check for NaN in Input Data
def check_for_nan(tensor, name):
    if torch.isnan(tensor).any() or torch.isinf(tensor).any():
        raise ValueError(f"{name} contains NaN or inf values: {tensor}")

# Verify input data
check_for_nan(data.x, "data.x")
check_for_nan(data.y, "data.y")
check_for_nan(data.edge_index, "data.edge_index")

# 2. Define a Simpler GNN Model with Task Grouping
class GroupedGNN(nn.Module):
    def __init__(self, num_features, hidden_dim, num_groups, task_groups, num_diseases):
        super(GroupedGNN, self).__init__()
        self.num_groups = num_groups
        self.task_groups = task_groups
        self.num_diseases = num_diseases

        self.conv1 = GCNConv(num_features, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, hidden_dim)

        self.group_heads = nn.ModuleList([
            nn.Linear(hidden_dim, len(group)) for group in task_groups
        ])

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        x = F.relu(x)

        outputs = []
        for group_idx, group in enumerate(self.task_groups):
            group_output = self.group_heads[group_idx](x)
            outputs.append(group_output)

        output = torch.zeros(x.size(0), self.num_diseases, device=x.device)
        for group_idx, group in enumerate(self.task_groups):
            for i, disease_idx in enumerate(group):
                output[:, disease_idx] = outputs[group_idx][:, i]

        if torch.isnan(output).any() or torch.isinf(output).any():
            raise ValueError("Model output contains NaN or inf before sigmoid")

        return torch.sigmoid(output)

# 3. Prepare for Training
num_nodes = data.x.size(0)
train_ratio, val_ratio = 0.7, 0.15
train_size = int(train_ratio * num_nodes)
val_size = int(val_ratio * num_nodes)

perm = torch.randperm(num_nodes)
train_mask = torch.zeros(num_nodes, dtype=torch.bool)
val_mask = torch.zeros(num_nodes, dtype=torch.bool)
test_mask = torch.zeros(num_nodes, dtype=torch.bool)

train_mask[perm[:train_size]] = 1
val_mask[perm[train_size:train_size + val_size]] = 1
test_mask[perm[train_size + val_size:]] = 1

data.train_mask = train_mask
data.val_mask = val_mask
data.test_mask = test_mask

# Compute class weights
pos_weights = []
for disease_idx in range(data.y.size(1)):
    pos_count = data.y[:, disease_idx].sum().item()
    neg_count = num_nodes - pos_count
    if pos_count > 0:
        weight = neg_count / pos_count
    else:
        weight = 1.0
    pos_weights.append(weight)

pos_weights = torch.tensor(pos_weights, dtype=torch.float32)
check_for_nan(pos_weights, "pos_weights")

# 4. Initialize the Model and Optimizer
hidden_dim = 16
model = GroupedGNN(
    num_features=data.x.size(1),
    hidden_dim=hidden_dim,
    num_groups=data.num_groups,
    task_groups=data.task_groups,
    num_diseases=data.y.size(1)
)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=5e-4)

# 5. Training Loop with Node Batching
def train():
    model.train()
    total_loss = 0
    train_indices = torch.where(data.train_mask)[0]
    perm = torch.randperm(len(train_indices))
    train_indices = train_indices[perm]

    batch_size = 2
    for start in range(0, len(train_indices), batch_size):
        end = min(start + batch_size, len(train_indices))
        batch_indices = train_indices[start:end]

        optimizer.zero_grad()
        out = model(data.x, data.edge_index)
        check_for_nan(out, "Model output in train")

        batch_out = out[batch_indices]
        batch_y = data.y[batch_indices]
        loss = F.binary_cross_entropy_with_logits(
            batch_out,
            batch_y,
            pos_weight=pos_weights
        )
        if torch.isnan(loss):
            raise ValueError("Loss is NaN")

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item() * len(batch_indices)

        del batch_out, batch_y, loss
        gc.collect()

    return total_loss / train_indices.size(0)

def evaluate(mask):
    model.eval()
    auroc_scores = []
    f1_scores = []
    indices = torch.where(mask)[0]

    batch_size = 2
    with torch.no_grad():
        for start in range(0, len(indices), batch_size):
            end = min(start + batch_size, len(indices))
            batch_indices = indices[start:end]

            out = model(data.x, data.edge_index)
            check_for_nan(out, "Model output in evaluate")

            preds = (out > 0.5).float()
            y_true = data.y[batch_indices].cpu().numpy()
            y_pred = out[batch_indices].cpu().numpy()
            y_pred_binary = preds[batch_indices].cpu().numpy()

            y_pred = np.nan_to_num(y_pred, nan=0.0)
            y_pred_binary = np.nan_to_num(y_pred_binary, nan=0.0)

            for disease_idx in range(data.y.size(1)):
                if y_true[:, disease_idx].sum() > 0:
                    auroc = roc_auc_score(y_true[:, disease_idx], y_pred[:, disease_idx])
                    f1 = f1_score(y_true[:, disease_idx], y_pred_binary[:, disease_idx])
                    auroc_scores.append(auroc)
                    f1_scores.append(f1)

            del out, preds, y_true, y_pred, y_pred_binary
            gc.collect()

    return (np.mean(auroc_scores) if auroc_scores else 0,
            np.mean(f1_scores) if f1_scores else 0)

# Training loop
num_epochs = 5  # Reduced to 5 epochs for faster testing
for epoch in range(num_epochs):
    train_loss = train()
    val_auroc, val_f1 = evaluate(data.val_mask)
    print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Val AUROC: {val_auroc:.4f}, Val F1: {val_f1:.4f}")

# 6. Evaluate on Test Set
test_auroc, test_f1 = evaluate(data.test_mask)
print(f"Test AUROC: {test_auroc:.4f}, Test F1: {test_f1:.4f}")

# Free up memory
del pos_weights, model, optimizer
gc.collect()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from torch_geometric.utils import subgraph
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score
import numpy as np
import gc
import time
import os

# Step 5: Continue Training the GNN Model with Task Grouping (Optimized)

# 1. Check for NaN in Input Data
def check_for_nan(tensor, name):
    if torch.isnan(tensor).any() or torch.isinf(tensor).any():
        raise ValueError(f"{name} contains NaN or inf values: {tensor}")

# Verify input data
check_for_nan(data.x, "data.x")
check_for_nan(data.y, "data.y")
check_for_nan(data.edge_index, "data.edge_index")

# Print number of diseases and edges for debugging
print(f"Number of diseases: {data.y.size(1)}")
print(f"Number of edges: {data.edge_index.shape[1]}")

# 2. Check Device (GPU if available)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Move data to device
data = data.to(device)

# 3. Define a Simpler GNN Model with Task Grouping
class GroupedGNN(nn.Module):
    def __init__(self, num_features, hidden_dim, num_groups, task_groups, num_diseases):
        super(GroupedGNN, self).__init__()
        self.num_groups = num_groups
        self.task_groups = task_groups
        self.num_diseases = num_diseases

        self.conv1 = GCNConv(num_features, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, hidden_dim)

        self.group_heads = nn.ModuleList([
            nn.Linear(hidden_dim, len(group)) for group in task_groups
        ])

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        x = F.relu(x)

        outputs = []
        for group_idx, group in enumerate(self.task_groups):
            group_output = self.group_heads[group_idx](x)
            outputs.append(group_output)

        output = torch.zeros(x.size(0), self.num_diseases, device=x.device)
        for group_idx, group in enumerate(self.task_groups):
            for i, disease_idx in enumerate(group):
                output[:, disease_idx] = outputs[group_idx][:, i]

        if torch.isnan(output).any() or torch.isinf(output).any():
            raise ValueError("Model output contains NaN or inf before sigmoid")

        return torch.sigmoid(output)

# 4. Prepare for Training
num_nodes = data.x.size(0)
train_ratio, val_ratio = 0.7, 0.15
train_size = int(train_ratio * num_nodes)
val_size = int(val_ratio * num_nodes)

perm = torch.randperm(num_nodes)
train_mask = torch.zeros(num_nodes, dtype=torch.bool)
val_mask = torch.zeros(num_nodes, dtype=torch.bool)
test_mask = torch.zeros(num_nodes, dtype=torch.bool)

train_mask[perm[:train_size]] = 1
val_mask[perm[train_size:train_size + val_size]] = 1
test_mask[perm[train_size + val_size:]] = 1

data.train_mask = train_mask
data.val_mask = val_mask
data.test_mask = test_mask

# Compute class weights
pos_weights = []
for disease_idx in range(data.y.size(1)):
    pos_count = data.y[:, disease_idx].sum().item()
    neg_count = num_nodes - pos_count
    if pos_count > 0:
        weight = neg_count / pos_count
    else:
        weight = 1.0
    pos_weights.append(weight)

pos_weights = torch.tensor(pos_weights, dtype=torch.float32).to(device)
check_for_nan(pos_weights, "pos_weights")

# 5. Initialize the Model and Optimizer
hidden_dim = 16
model = GroupedGNN(
    num_features=data.x.size(1),
    hidden_dim=hidden_dim,
    num_groups=data.num_groups,
    task_groups=data.task_groups,
    num_diseases=data.y.size(1)
).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4)  # Increased learning rate

# 6. Checkpointing Functions
def save_checkpoint(model, optimizer, epoch, filename='checkpoint.pth'):
    checkpoint = {
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'epoch': epoch
    }
    torch.save(checkpoint, filename)
    print(f"Saved checkpoint to {filename}")

def load_checkpoint(model, optimizer, filename='checkpoint.pth'):
    if os.path.exists(filename):
        checkpoint = torch.load(filename, map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        start_epoch = checkpoint['epoch'] + 1
        print(f"Loaded checkpoint from {filename}, starting from epoch {start_epoch}")
        return start_epoch
    else:
        print(f"No checkpoint found at {filename}, starting from epoch 1")
        return 1

# 7. Training and Evaluation Functions
def train(model, optimizer, data, batch_size=8):
    model.train()
    total_loss = 0
    train_indices = torch.where(data.train_mask)[0]
    perm = torch.randperm(len(train_indices))
    train_indices = train_indices[perm]

    start_time = time.time()
    for start in range(0, len(train_indices), batch_size):
        batch_start_time = time.time()
        end = min(start + batch_size, len(train_indices))
        batch_indices = train_indices[start:end]

        # Extract subgraph for the batch
        subset = batch_indices.to(device)
        edge_index, _ = subgraph(subset, data.edge_index, relabel_nodes=True, num_nodes=num_nodes)
        x_batch = data.x[subset].to(device)
        edge_index = edge_index.to(device)

        optimizer.zero_grad()
        out = model(x_batch, edge_index)
        check_for_nan(out, "Model output in train")

        batch_y = data.y[subset]
        loss = F.binary_cross_entropy_with_logits(
            out,
            batch_y,
            pos_weight=pos_weights
        )
        if torch.isnan(loss):
            raise ValueError("Loss is NaN")

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item() * len(batch_indices)

        del out, batch_y, loss, x_batch, edge_index
        gc.collect()

        batch_time = time.time() - batch_start_time
        print(f"Batch {start//batch_size + 1}, Time: {batch_time:.2f}s")

    epoch_time = time.time() - start_time
    print(f"Train Epoch Time: {epoch_time:.2f}s")
    return total_loss / train_indices.size(0)

def evaluate(model, data, mask, compute_metrics=True):
    model.eval()
    indices = torch.where(mask)[0]

    start_time = time.time()
    with torch.no_grad():
        subset = indices.to(device)
        edge_index, _ = subgraph(subset, data.edge_index, relabel_nodes=True, num_nodes=num_nodes)
        x_subset = data.x[subset].to(device)
        edge_index = edge_index.to(device)

        out = model(x_subset, edge_index)
        check_for_nan(out, "Model output in evaluate")

        if compute_metrics:
            preds = (out > 0.3).float()  # Threshold for binary predictions
            y_true = data.y[subset].cpu().numpy()
            y_pred = out.cpu().numpy()
            y_pred_binary = preds.cpu().numpy()

            y_pred = np.nan_to_num(y_pred, nan=0.0)
            y_pred_binary = np.nan_to_num(y_pred_binary, nan=0.0)

            pos_counts = y_true.sum(axis=0)
            neg_counts = len(y_true) - pos_counts
            valid_diseases = (pos_counts > 0) & (neg_counts > 0)

            auroc_scores = []
            f1_scores = []
            accuracy_scores = []
            for disease_idx in range(data.y.size(1)):
                if valid_diseases[disease_idx]:
                    auroc = roc_auc_score(y_true[:, disease_idx], y_pred[:, disease_idx])
                    f1 = f1_score(y_true[:, disease_idx], y_pred_binary[:, disease_idx])
                    accuracy = accuracy_score(y_true[:, disease_idx], y_pred_binary[:, disease_idx])
                    auroc_scores.append(auroc)
                    f1_scores.append(f1)
                    accuracy_scores.append(accuracy)

            del preds, y_true, y_pred, y_pred_binary
        else:
            auroc_scores = []
            f1_scores = []
            accuracy_scores = []

        del out, x_subset, edge_index
        gc.collect()

    eval_time = time.time() - start_time
    print(f"Evaluate Time: {eval_time:.2f}s")
    return (np.mean(auroc_scores) if auroc_scores else 0,
            np.mean(f1_scores) if f1_scores else 0,
            np.mean(accuracy_scores) if accuracy_scores else 0)

# 8. Incremental Training Loop
checkpoint_file = 'gnn_checkpoint.pth'
start_epoch = load_checkpoint(model, optimizer, checkpoint_file)
total_epochs = 50  # Target total epochs
epochs_per_session = 10  # Train 10 more epochs in this session
end_epoch = min(start_epoch + epochs_per_session - 1, total_epochs)

print(f"Training from epoch {start_epoch} to {end_epoch}...")

total_start_time = time.time()
for epoch in range(start_epoch, end_epoch + 1):
    epoch_start_time = time.time()
    train_loss = train(model, optimizer, data, batch_size=8)
    compute_metrics = (epoch % 5 == 0) or (epoch == total_epochs)  # Compute metrics every 5 epochs or at the end
    val_auroc, val_f1, val_accuracy = evaluate(model, data, data.val_mask, compute_metrics=compute_metrics)
    epoch_time = time.time() - epoch_start_time
    if compute_metrics:
        print(f"Epoch {epoch}/{total_epochs}, Train Loss: {train_loss:.4f}, Val AUROC: {val_auroc:.4f}, Val F1: {val_f1:.4f}, Val Accuracy: {val_accuracy:.4f}, Epoch Time: {epoch_time:.2f}s")
    else:
        print(f"Epoch {epoch}/{total_epochs}, Train Loss: {train_loss:.4f}, Epoch Time: {epoch_time:.2f}s")

    # Save checkpoint after each epoch
    save_checkpoint(model, optimizer, epoch, checkpoint_file)

# 9. Evaluate on Test Set at Current Epoch (Epoch 20)
print(f"\nEvaluating on Test Set at Epoch {end_epoch}...")
test_auroc, test_f1, test_accuracy = evaluate(model, data, data.test_mask, compute_metrics=True)
print(f"Test AUROC: {test_auroc:.4f}, Test F1: {test_f1:.4f}, Test Accuracy: {test_accuracy:.4f}")

# 10. Check if Training is Complete
if end_epoch == total_epochs:
    print(f"Training complete at epoch {end_epoch}.")
else:
    print(f"Training not complete. Current epoch: {end_epoch}. Run again to continue training.")

total_time = time.time() - total_start_time
print(f"Total Training Time for this session: {total_time:.2f}s")

# Free up memory
# Do not delete pos_weights, model, or optimizer yet, as we'll need them for Step 6
gc.collect()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from torch_geometric.utils import subgraph
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score
import numpy as np
import gc
import time
import os

# Step 5: Continue Training the GNN Model with Task Grouping (Optimized)

# 1. Check for NaN in Input Data
def check_for_nan(tensor, name):
    if torch.isnan(tensor).any() or torch.isinf(tensor).any():
        raise ValueError(f"{name} contains NaN or inf values: {tensor}")

# Verify input data
check_for_nan(data.x, "data.x")
check_for_nan(data.y, "data.y")
check_for_nan(data.edge_index, "data.edge_index")

# Print number of diseases and edges for debugging
print(f"Number of diseases: {data.y.size(1)}")
print(f"Number of edges: {data.edge_index.shape[1]}")

# 2. Check Device (GPU if available)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Move data to device
data = data.to(device)

# 3. Define a Simpler GNN Model with Task Grouping
class GroupedGNN(nn.Module):
    def __init__(self, num_features, hidden_dim, num_groups, task_groups, num_diseases):
        super(GroupedGNN, self).__init__()
        self.num_groups = num_groups
        self.task_groups = task_groups
        self.num_diseases = num_diseases

        self.conv1 = GCNConv(num_features, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, hidden_dim)

        self.group_heads = nn.ModuleList([
            nn.Linear(hidden_dim, len(group)) for group in task_groups
        ])

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        x = F.relu(x)

        outputs = []
        for group_idx, group in enumerate(self.task_groups):
            group_output = self.group_heads[group_idx](x)
            outputs.append(group_output)

        output = torch.zeros(x.size(0), self.num_diseases, device=x.device)
        for group_idx, group in enumerate(self.task_groups):
            for i, disease_idx in enumerate(group):
                output[:, disease_idx] = outputs[group_idx][:, i]

        if torch.isnan(output).any() or torch.isinf(output).any():
            raise ValueError("Model output contains NaN or inf before sigmoid")

        return torch.sigmoid(output)

# 4. Prepare for Training
num_nodes = data.x.size(0)
train_ratio, val_ratio = 0.7, 0.15
train_size = int(train_ratio * num_nodes)
val_size = int(val_ratio * num_nodes)

perm = torch.randperm(num_nodes)
train_mask = torch.zeros(num_nodes, dtype=torch.bool)
val_mask = torch.zeros(num_nodes, dtype=torch.bool)
test_mask = torch.zeros(num_nodes, dtype=torch.bool)

train_mask[perm[:train_size]] = 1
val_mask[perm[train_size:train_size + val_size]] = 1
test_mask[perm[train_size + val_size:]] = 1

data.train_mask = train_mask
data.val_mask = val_mask
data.test_mask = test_mask

# Compute class weights
pos_weights = []
for disease_idx in range(data.y.size(1)):
    pos_count = data.y[:, disease_idx].sum().item()
    neg_count = num_nodes - pos_count
    if pos_count > 0:
        weight = neg_count / pos_count
    else:
        weight = 1.0
    pos_weights.append(weight)

pos_weights = torch.tensor(pos_weights, dtype=torch.float32).to(device)
check_for_nan(pos_weights, "pos_weights")

# 5. Initialize the Model and Optimizer
hidden_dim = 16
model = GroupedGNN(
    num_features=data.x.size(1),
    hidden_dim=hidden_dim,
    num_groups=data.num_groups,
    task_groups=data.task_groups,
    num_diseases=data.y.size(1)
).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4)

# 6. Checkpointing Functions
def save_checkpoint(model, optimizer, epoch, filename='checkpoint.pth'):
    checkpoint = {
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'epoch': epoch
    }
    torch.save(checkpoint, filename)
    print(f"Saved checkpoint to {filename}")

def load_checkpoint(model, optimizer, filename='checkpoint.pth'):
    if os.path.exists(filename):
        checkpoint = torch.load(filename, map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        start_epoch = checkpoint['epoch'] + 1
        print(f"Loaded checkpoint from {filename}, starting from epoch {start_epoch}")
        return start_epoch
    else:
        print(f"No checkpoint found at {filename}, starting from epoch 1")
        return 1

# 7. Training and Evaluation Functions
def train(model, optimizer, data, batches):
    model.train()
    total_loss = 0

    start_time = time.time()
    for batch_idx, (subset, edge_index) in enumerate(batches):
        batch_start_time = time.time()

        optimizer.zero_grad()
        out = model(data.x[subset], edge_index)
        check_for_nan(out, "Model output in train")

        batch_y = data.y[subset]
        loss = F.binary_cross_entropy_with_logits(
            out,
            batch_y,
            pos_weight=pos_weights
        )
        if torch.isnan(loss):
            raise ValueError("Loss is NaN")

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item() * len(subset)

        del out, batch_y, loss

        batch_time = time.time() - batch_start_time
        print(f"Batch {batch_idx + 1}, Time: {batch_time:.2f}s")

    epoch_time = time.time() - start_time
    print(f"Train Epoch Time: {epoch_time:.2f}s")
    return total_loss / len(batches)

def evaluate(model, data, mask, compute_metrics=True):
    model.eval()
    indices = torch.where(mask)[0]

    start_time = time.time()
    with torch.no_grad():
        subset = indices.to(device)
        edge_index, _ = subgraph(subset, data.edge_index, relabel_nodes=True, num_nodes=num_nodes)
        x_subset = data.x[subset].to(device)
        edge_index = edge_index.to(device)

        out = model(x_subset, edge_index)
        check_for_nan(out, "Model output in evaluate")

        if compute_metrics:
            preds = (out > 0.3).float()
            y_true = data.y[subset].cpu().numpy()
            y_pred = out.cpu().numpy()
            y_pred_binary = preds.cpu().numpy()

            y_pred = np.nan_to_num(y_pred, nan=0.0)
            y_pred_binary = np.nan_to_num(y_pred_binary, nan=0.0)

            pos_counts = y_true.sum(axis=0)
            neg_counts = len(y_true) - pos_counts
            valid_diseases = (pos_counts > 0) & (neg_counts > 0)

            auroc_scores = []
            f1_scores = []
            accuracy_scores = []
            for disease_idx in range(data.y.size(1)):
                if valid_diseases[disease_idx]:
                    auroc = roc_auc_score(y_true[:, disease_idx], y_pred[:, disease_idx])
                    f1 = f1_score(y_true[:, disease_idx], y_pred_binary[:, disease_idx])
                    accuracy = accuracy_score(y_true[:, disease_idx], y_pred_binary[:, disease_idx])
                    auroc_scores.append(auroc)
                    f1_scores.append(f1)
                    accuracy_scores.append(accuracy)

            del preds, y_true, y_pred, y_pred_binary
        else:
            auroc_scores = []
            f1_scores = []
            accuracy_scores = []

        del out, x_subset, edge_index

    eval_time = time.time() - start_time
    print(f"Evaluate Time: {eval_time:.2f}s")
    return (np.mean(auroc_scores) if auroc_scores else 0,
            np.mean(f1_scores) if f1_scores else 0,
            np.mean(accuracy_scores) if accuracy_scores else 0)

# 8. Precompute Subgraphs for Training Batches
train_indices = torch.where(data.train_mask)[0]
batch_size = 32
batches = []
for start in range(0, len(train_indices), batch_size):
    end = min(start + batch_size, len(train_indices))
    batch_indices = train_indices[start:end]
    subset = batch_indices.to(device)
    edge_index, _ = subgraph(subset, data.edge_index, relabel_nodes=True, num_nodes=num_nodes)
    batches.append((subset, edge_index))

# 9. Incremental Training Loop
checkpoint_file = 'gnn_checkpoint.pth'
start_epoch = load_checkpoint(model, optimizer, checkpoint_file)
total_epochs = 50
epochs_per_session = 10
end_epoch = min(start_epoch + epochs_per_session - 1, total_epochs)

print(f"Training from epoch {start_epoch} to {end_epoch}...")

total_start_time = time.time()
for epoch in range(start_epoch, end_epoch + 1):
    epoch_start_time = time.time()
    train_loss = train(model, optimizer, data, batches)
    compute_metrics = (epoch % 5 == 0) or (epoch == total_epochs)
    val_auroc, val_f1, val_accuracy = evaluate(model, data, data.val_mask, compute_metrics=compute_metrics)
    epoch_time = time.time() - epoch_start_time
    if compute_metrics:
        print(f"Epoch {epoch}/{total_epochs}, Train Loss: {train_loss:.4f}, Val AUROC: {val_auroc:.4f}, Val F1: {val_f1:.4f}, Val Accuracy: {val_accuracy * 100:.2f}%, Epoch Time: {epoch_time:.2f}s")
    else:
        print(f"Epoch {epoch}/{total_epochs}, Train Loss: {train_loss:.4f}, Epoch Time: {epoch_time:.2f}s")

    save_checkpoint(model, optimizer, epoch, checkpoint_file)
    gc.collect()

# 10. Evaluate on Test Set at Current Epoch
print(f"\nEvaluating on Test Set at Epoch {end_epoch}...")
test_auroc, test_f1, test_accuracy = evaluate(model, data, data.test_mask, compute_metrics=True)
print(f"Test AUROC: {test_auroc:.4f}, Test F1: {test_f1:.4f}, Test Accuracy: {test_accuracy * 100:.2f}%")

# 11. Check if Training is Complete
if end_epoch == total_epochs:
    print(f"Training complete at epoch {end_epoch}.")
else:
    print(f"Training not complete. Current epoch: {end_epoch}. Run again to continue training.")

total_time = time.time() - total_start_time
print(f"Total Training Time for this session: {total_time:.2f}s")

# 12. Preserve variables for Step 6
# pos_weights, model, and optimizer are already preserved
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.utils import subgraph
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
import gc
import time
import os

# Step 6: Analyze Results and Fine-Tune

# 1. Check Device (Ensure consistency with Step 5)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Ensure data, model, and pos_weights are on the correct device
data = data.to(device)
model = model.to(device)
pos_weights = pos_weights.to(device)

# 2. Per-Disease Performance Analysis
def per_disease_analysis(model, data, mask, label_cols):
    model.eval()
    indices = torch.where(mask)[0]

    with torch.no_grad():
        subset = indices.to(device)
        edge_index, _ = subgraph(subset, data.edge_index, relabel_nodes=True, num_nodes=data.x.size(0))
        x_subset = data.x[subset].to(device)
        edge_index = edge_index.to(device)

        out = model(x_subset, edge_index)
        preds = (out > 0.3).float()
        y_true = data.y[subset].cpu().numpy()
        y_pred = out.cpu().numpy()
        y_pred_binary = preds.cpu().numpy()

        y_pred = np.nan_to_num(y_pred, nan=0.0)
        y_pred_binary = np.nan_to_num(y_pred_binary, nan=0.0)

        pos_counts = y_true.sum(axis=0)
        neg_counts = len(y_true) - pos_counts
        valid_diseases = (pos_counts > 0) & (neg_counts > 0)

        results = []
        for disease_idx in range(data.y.size(1)):
            if valid_diseases[disease_idx]:
                auroc = roc_auc_score(y_true[:, disease_idx], y_pred[:, disease_idx])
                f1 = f1_score(y_true[:, disease_idx], y_pred_binary[:, disease_idx])
                accuracy = accuracy_score(y_true[:, disease_idx], y_pred_binary[:, disease_idx])
                results.append({
                    'Disease': label_cols[disease_idx],
                    'AUROC': auroc,
                    'F1': f1,
                    'Accuracy (%)': accuracy * 100,  # Convert to percentage for storage
                    'Positive Count': pos_counts[disease_idx],
                    'Negative Count': neg_counts[disease_idx]
                })

        del preds, y_true, y_pred, y_pred_binary
        gc.collect()

    return pd.DataFrame(results)

# Get label_cols from Step 2
# If not available, recreate it as:
# label_cols = [col for col in patient_data_sub.columns if col.startswith('Label_')]

# Analyze performance on test set
test_results = per_disease_analysis(model, data, data.test_mask, label_cols)
print("\nPer-Disease Performance on Test Set:")
# Format the DataFrame display to show Accuracy as percentage
print(test_results.to_string(formatters={'Accuracy (%)': '{:.2f}%'.format}))

# Save to CSV
test_results.to_csv('per_disease_performance.csv', index=False)
print("Saved per-disease performance to 'per_disease_performance.csv'")

# 3. Evaluate Task Group Effectiveness
def evaluate_task_groups(model, data, mask, task_groups):
    model.eval()
    indices = torch.where(mask)[0]

    with torch.no_grad():
        subset = indices.to(device)
        edge_index, _ = subgraph(subset, data.edge_index, relabel_nodes=True, num_nodes=data.x.size(0))
        x_subset = data.x[subset].to(device)
        edge_index = edge_index.to(device)

        out = model(x_subset, edge_index)
        y_true = data.y[subset].cpu().numpy()
        y_pred = out.cpu().numpy()

        y_pred = np.nan_to_num(y_pred, nan=0.0)

        group_aurocs = []
        for group_idx, group in enumerate(task_groups):
            group_indices = list(group)
            pos_counts = y_true[:, group_indices].sum(axis=0)
            neg_counts = len(y_true) - pos_counts
            valid_diseases = (pos_counts > 0) & (neg_counts > 0)

            auroc_scores = []
            for idx, disease_idx in enumerate(group_indices):
                if valid_diseases[idx]:
                    auroc = roc_auc_score(y_true[:, disease_idx], y_pred[:, disease_idx])
                    auroc_scores.append(auroc)
            group_auroc = np.mean(auroc_scores) if auroc_scores else 0
            group_aurocs.append(group_auroc)
            print(f"Task Group {group_idx + 1} AUROC: {group_auroc:.4f}")

    return group_aurocs

# Evaluate current task groups
group_aurocs = evaluate_task_groups(model, data, data.test_mask, data.task_groups)
print(f"Average AUROC across task groups: {np.mean(group_aurocs):.4f}")

# 4. Fine-Tune Task Grouping with Agglomerative Clustering
# Compute co-occurrence matrix (same as in Step 4)
y_np = data.y.cpu().numpy()
co_occurrence = np.zeros((data.y.size(1), data.y.size(1)))
for i in range(data.y.size(1)):
    for j in range(data.y.size(1)):
        if i != j:
            co_occurrence[i, j] = np.sum((y_np[:, i] == 1) & (y_np[:, j] == 1))
co_occurrence = co_occurrence / (y_np.shape[0] * 2)
similarity_matrix = co_occurrence + co_occurrence.T

# Apply Agglomerative Clustering
num_groups = data.num_groups
# Updated to use 'metric' instead of 'affinity'
clustering = AgglomerativeClustering(n_clusters=num_groups, metric='precomputed', linkage='average')
labels = clustering.fit_predict(1 - similarity_matrix)

# Create new task groups
new_task_groups = [[] for _ in range(num_groups)]
for disease_idx, cluster_id in enumerate(labels):
    new_task_groups[cluster_id].append(disease_idx)

# Convert to list of indices
new_task_groups = [torch.tensor(group, dtype=torch.long) for group in new_task_groups]
data.task_groups = new_task_groups
print("Updated task groups with Agglomerative Clustering:")
for i, group in enumerate(new_task_groups):
    print(f"Group {i + 1}: {group.tolist()}")

# 5. Reinitialize Model with New Task Groups
hidden_dim = 16
model = GroupedGNN(
    num_features=data.x.size(1),
    hidden_dim=hidden_dim,
    num_groups=num_groups,
    task_groups=new_task_groups,
    num_diseases=data.y.size(1)
).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4)

# 6. Train with New Task Groups
num_epochs = 10
batch_size = 32

# Precompute subgraphs
train_indices = torch.where(data.train_mask)[0]
batches = []
for start in range(0, len(train_indices), batch_size):
    end = min(start + batch_size, len(train_indices))
    batch_indices = train_indices[start:end]
    subset = batch_indices.to(device)
    edge_index, _ = subgraph(subset, data.edge_index, relabel_nodes=True, num_nodes=data.x.size(0))
    batches.append((subset, edge_index))

# Training loop
print(f"\nFine-tuning with new task groups for {num_epochs} epochs...")
total_start_time = time.time()
for epoch in range(1, num_epochs + 1):
    epoch_start_time = time.time()
    train_loss = train(model, optimizer, data, batches)
    val_auroc, val_f1, val_accuracy = evaluate(model, data, data.val_mask, compute_metrics=True)
    epoch_time = time.time() - epoch_start_time
    print(f"Epoch {epoch}/{num_epochs}, Train Loss: {train_loss:.4f}, Val AUROC: {val_auroc:.4f}, Val F1: {val_f1:.4f}, Val Accuracy: {val_accuracy * 100:.2f}%, Epoch Time: {epoch_time:.2f}s")
    gc.collect()

# 7. Final Evaluation on Test Set
test_auroc, test_f1, test_accuracy = evaluate(model, data, data.test_mask, compute_metrics=True)
print(f"\nFinal Test AUROC: {test_auroc:.4f}, Test F1: {test_f1:.4f}, Test Accuracy: {test_accuracy * 100:.2f}%")

# 8. Save Final Results
final_results = {
    'Test AUROC': test_auroc,
    'Test F1': test_f1,
    'Test Accuracy (%)': test_accuracy * 100  # Store as percentage
}
pd.DataFrame([final_results]).to_csv('final_results.csv', index=False)
print("Saved final results to 'final_results.csv'")

total_time = time.time() - total_start_time
print(f"Total Fine-Tuning Time: {total_time:.2f}s")

# 9. Clean Up
gc.collect()
```

```python
# Step 5: GNN Training

import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
import os

# Define the GroupedGNN model class (same as Step 6)
class GroupedGNN(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, num_labels, num_groups):
        super(GroupedGNN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, hidden_channels)
        # Debug: Print group sizes
        group_sizes = [(data.task_groups == g).sum().item() for g in range(num_groups)]
        print(f"Group sizes for heads: {group_sizes}")
        if any(size == 0 for size in group_sizes):
            raise ValueError("One or more task groups have no labels assigned. Check task_groups.")
        self.heads = torch.nn.ModuleList([
            torch.nn.Linear(hidden_channels, (data.task_groups == g).sum().item())
            for g in range(num_groups)
        ])

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        outputs = []
        for head in self.heads:
            group_out = head(x)
            outputs.append(group_out)
        return torch.cat(outputs, dim=1)

# Initialize the model
model = GroupedGNN(
    in_channels=data.x.shape[1],
    hidden_channels=64,
    num_labels=data.y.shape[1],
    num_groups=data.num_groups
)

# Set up optimizer and loss
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
pos_weight = torch.ones(data.y.shape[1]) * 5  # Adjust for class imbalance

# Training loop
model.train()
for epoch in range(50):
    optimizer.zero_grad()
    out = model(data)
    loss = F.binary_cross_entropy_with_logits(
        out[data.train_mask], data.y[data.train_mask], pos_weight=pos_weight
    )
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        # Evaluate on validation set
        model.eval()
        with torch.no_grad():
            pred = torch.sigmoid(out[data.val_mask])
            pred_binary = (pred > 0.5).float()
            accuracy = (pred_binary == data.y[data.val_mask]).float().mean().item() * 100
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}, Validation Accuracy: {accuracy:.2f}%")
        model.train()

# Save the model
model_path = '/content/drive/My Drive/GNN_Project/gnn_model.pth'
torch.save(model.state_dict(), model_path)
print(f"Saved model to {model_path}")

# Verify the model was saved
if os.path.exists(model_path):
    print(f"Confirmed model exists at {model_path}")
else:
    raise FileNotFoundError(f"Failed to save model at {model_path}")
```
